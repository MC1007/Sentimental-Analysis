```python
Polarity_Score_list = []
Subjectivity_Score_list= []
positive_word_count_list = []
negative_word_count_list = []
Average_Sentence_Length_list = []
Percentage_Complex_words_list = []
Fog_Index_list = []
avg_words_per_sentence_list = []
complex_words_count_list = []
Word_Count_list = []

Avg_Word_Length_list = []

for urls in df['URL']:
        # Send a request to the webpage and store the response as a variable
        r = requests.get(urls)
        htmlcontent = r.content
        #print(htmlcontent)
        soup =  BeautifulSoup(htmlcontent,'html.parser')
        #print(soup)
        # Extract all text from the webpage
        text = soup.get_text()
        #print (text)
        # Remove stop words from the extracted text
        stop_words = set(stopwords.words('english'))
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in stop_words]
        filtered_text = ' '.join(filtered_words)
        # Tokenize the input statement into words
        statement = filtered_text
        words = nltk.word_tokenize(statement)
        sentences = nltk.sent_tokenize(statement)
        num_sentences = len(sentences)
        total_words = len(filtered_words) # total words after cleaning
        Word_Count_list.append(total_words)
        #print(total_words)
        # Initialize a counter variable
        positive_word_count = 0
        negative_word_count = 0
        # Iterate over each word and count positive words
        for word in words:
            if word in Positive_words_list:
                positive_word_count += 1
            if word in negative_words_list:
                negative_word_count +=1
        Polarity_Score = (positive_word_count - negative_word_count)/ ((positive_word_count + negative_word_count) + 0.000001)
        Polarity_Score_list.append(Polarity_Score)
        Subjectivity_Score = (positive_word_count + negative_word_count)/ ((total_words) + 0.000001)
        Subjectivity_Score_list.append(Subjectivity_Score)
        positive_word_count_list.append(positive_word_count)
        negative_word_count_list.append(negative_word_count)
        num_complex_words = 0
        for word in words:
            num_syllables = syllables.estimate(word)
            if num_syllables > 2:
                num_complex_words += 1
        print(num_complex_words)
        complex_words_count_list.append(num_complex_words)
        Average_Sentence_Length = total_words / num_sentences
        Average_Sentence_Length_list.append(Average_Sentence_Length)
        Percentage_Complex_words = num_complex_words / total_words
        Percentage_Complex_words_list.append(Percentage_Complex_words)
        Fog_Index = 0.4 * (Average_Sentence_Length + Percentage_Complex_words)
        Fog_Index_list.append(Fog_Index)
        avg_words_per_sentence = total_words / len(sentences)
        avg_words_per_sentence_list.append(avg_words_per_sentence)

        print(positive_word_count)
        print(negative_word_count)
'''print('Average_Sentence_Length',Average_Sentence_Length)
        print('Percentage_Complex_words',Percentage_Complex_words)
        print('Fog_Index',Fog_Index)'''
      
```

```python
vowels = "aeiouy" # Define vowels
exceptions = ["es", "ed"] # Define exceptions
syellable_per_word_list = []
for urls in df['URL']:
        # Send a request to the webpage and store the response as a variable
        r = requests.get(urls)
        htmlcontent = r.content
        #print(htmlcontent)
        soup =  BeautifulSoup(htmlcontent,'html.parser')
        #print(soup)
        # Extract all text from the webpage
        text = soup.get_text()
        #print (text)
        # Remove stop words from the extracted text
        stop_words = set(stopwords.words('english'))
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in stop_words]
        filtered_text = ' '.join(filtered_words)
        vowels = "aeiouy" # Define vowels
        exceptions = ["es", "ed"] # Define exceptions

        syllable_count_per_word = 0

        for word in filtered_text.split():
            syllables_count = 0
            last_was_vowel = False
            for letter in word.lower():
                if letter in vowels:
                    if not last_was_vowel:
                        syllables_count += 1
                    last_was_vowel = True
                else:
                    last_was_vowel = False
            for exception in exceptions:
                if word.lower().endswith(exception):
                    syllables_count -= 1
            syllable_count_per_word += 1
        print("The total number of syllables in the text is:", syllable_count_per_word)
        syellable_per_word_list.append(syllable_count_per_word)
```
